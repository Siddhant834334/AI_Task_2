**Log transformation to reduce skewness**

```python
# Safe ratio with small epsilon to avoid division by zero
atlantis['Wealth_per_sqft'] = atlantis['Wealth_Index'] / (atlantis['House_Size_sq_ft'] + 1e-6)
atlantis_test['Wealth_per_sqft'] = atlantis_test['Wealth_Index'] / (atlantis_test['House_Size_sq_ft'] + 1e-6)

# Apply log transform consistently
log_cols = ['Wealth_Index', 'House_Size_sq_ft', 'Life_Expectancy', 'Wealth_per_sqft']
for col in log_cols:
    atlantis[col]=np.log1p(atlantis[col])
    atlantis_test[col] = np.log1p(atlantis_test[col])

```

```python
X=atlantis.drop(['Occupation','Bio_Hash','Citizen_ID'],axis=1)
y=atlantis['Occupation']
X_test_final=atlantis_test.drop(['Bio_Hash','Citizen_ID'],axis=1)
```

```python
trans_col=[ 'District_Name', 'Work_District','Diet_Type','Vehicle_Owned']
Oenc=OrdinalEncoder()
X[trans_col]=Oenc.fit_transform(X[trans_col])
X_test_final[trans_col]=Oenc.transform(X_test_final[trans_col])
```

```python
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
model=XGBClassifier(colsample_bytree=0.8, gamma= 9, learning_rate=0.05, max_depth= 6, n_estimators=200, reg_alpha=4, reg_lambda=6,random_state=42)
scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
print(scores.mean())
scores
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, GroupKFold
from sklearn.metrics import accuracy_score
groups=atlantis['District_Name']
# Define GroupKFold
gkf = GroupKFold(n_splits=4)
model = XGBClassifier(
    objective='multi:softprob',
    num_class=5,
    eval_metric='mlogloss',
    use_label_encoder=False,
    random_state=42
)
param_grid = {
    'n_estimators': [700],
    'max_depth': [7],
    'learning_rate': [0.055],
    'min_child_weight': [1],
    'subsample': [0.8],
    'colsample_bytree': [0.8],
    'gamma': [1],
    'reg_alpha': [3],
    'reg_lambda': [7]
}
# GridSearch with GroupKFold
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='accuracy',
    cv=gkf.split(X, y, groups),
    verbose=1,
    n_jobs=-1
)
grid_search.fit(X, y)
best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
best_model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=True)
y_pred_val = best_model.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_pred_val))
```
